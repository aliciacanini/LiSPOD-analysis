## 3. Processamento dei dati

Dall'esplorazione dei dati si evincono due problemi principali:

1. Tutte le 14 colonne sono di tipo `object`, nonostante alcuni valori siano numeri interi (quindi dovrebbero essere `int`)

2. Sono presenti numerosi dati mancanti *NaN*, specialmente nelle colonne di
 **argomenti** e **doi_disponibili**.  

*   **argomenti** (42%)
*   **doi_disponibili** (49%)

Significa che per 749 articoli non sappiamo di cosa parlino e solo la metà degli articoli ha un identificatore univoco.

Inizio gestendo i NaN delle colonne che contengono i valori da trasfomare in `int`.

Trovo i valori nulli nella colonna **data_pubblicazione**:

```python
df["data_pubblicazione"].notna()
```

```python
# creo un nuovo df senza valori nulli in data_pubblicazione
df_cronologia = df[df["data_pubblicazione"].notna()]
```

```python
df_cronologia
```

```python
df_cronologia['data_pubblicazione'] = df_cronologia['data_pubblicazione'].astype(str).str.split('|').str[0]
```

Ho aggiunto il metodo `.split()` perchè nella colonna **data_pubblicazione** erano presenti sei valori in cui la data era indicata come "1987 | 1989", perciò non sarebbe stato possibile trasformare i valori in `int`.

Trasformo i valori nella colonna da `object` a `int`.

```python
df_cronologia.data_pubblicazione = df_cronologia.data_pubblicazione.astype(int)
```

```python
df_cronologia.info()
```

Ora la colonna **data_pubblicazione** è un `int`.

In questa fase ho scelto di procedere alla conversione tipologica esclusivamente per la colonna **data_pubblicazione**, trasformandola in formato numerico `int` per permettere l'analisi cronologica.


Altre colonne come edizione, volume e pagine, sebbene presentino valori numerici, non sono state convertite e vengono mantenute nel formato originale `object` perchè:

1. Questi metadati non sono strettamente necessari ai fini degli obiettivi di questa analisi.

2. Il formato testuale permette di preservare informazioni complesse (es. intervalli di pagine come '35-37') che andrebbero perse con una conversione numerica.

Il passo successivo è gestire i valori NaN nelle colonne **argomenti** e **doi_disponibili**.

Scelgo di indicare i dati mancanti con `Non Specificato` ai fini della mia analisi e per evitare che vengano eliminate troppe righe.


```python
df_mancanti = df.copy()
df_mancanti["argomenti"] = df_mancanti["argomenti"].fillna("Non specificato")
```

```python
df_mancanti["doi_disponibili"] = df_mancanti["doi_disponibili"].fillna("Non specificato")
```

```python
df_mancanti
```

Un'altra analisi interessante da effettuare è quella sulla *qualità dei metadati*, ovvero sulla quantità di informazioni mancanti (**NaN**). Le due colonne con percentuale maggiore di argomenti mancanti sono **argomenti** e **doi_disponibili**.  

```python
percentuale_nan = df.isnull().mean() * 100
print(percentuale_nan)
```

```python
# definisco la funzione di base per il grafico
def argomenti_grafico(argomenti):
    if argomenti == "Non specificato" :
        return "Mancante"
    else:
        return "Presente"
```

```python
# applico la funzione
argomenti_pie = df_mancanti['argomenti'].apply(argomenti_grafico)
```

```python
ax = argomenti_pie.value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=['pink','grey'],
    figsize=(7, 7),
    ylabel='',
    startangle=90,
    explode=(0.05, 0),
    textprops={'fontsize': 14, 'color': 'black'},
    pctdistance=0.6
)

ax.set_title("Qualità dei metadati: Argomenti", fontsize=20)
```

```python
# faccio la stessa cosa con i DOI
def doi_grafico(doi_disponibili):
    if doi_disponibili == "Non specificato" :
        return "Mancante"
    else:
        return "Presente"
```

```python
doi_pie = df_mancanti['doi_disponibili'].apply(doi_grafico)
```

```python
ax = doi_pie.value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=['yellow','gray'],
    figsize=(7, 7),
    ylabel='',
    startangle=90,
    explode=(0.05, 0),
    textprops={'fontsize': 14, 'color': 'black'},
    pctdistance=0.6
)

ax.set_title("Qualità dei metadati: DOI", fontsize=20)
```

I grafici evidenziano, in entrambi i casi, una mancanza di dati imponente. Nonostante la mole di dati, la metà del dataset presenta lacune che rendono questi record "silenziosi".

### 4.2 Explanatory Data Analysis

Dopo aver effettuato un'analisi **cronologica** sulle date delle pubblicazioni e un'analisi **qualitativa** sui dati forniti da Wikidata, è chiara una convergenza sul piano della mancanza di dati.

Nonostante il dataset LiSPOD mostri una crescita numerica di record a partire dal 2010, questa crescita non è accompagnata da una uguale solidità dei metadati.

Questa convergenza nelle **lacune** è particolarmente evidente nell'ultimo quinquennio. Il calo drastico dei dati dal 2021 al 2025 non indica una riduzione della ricerca scientifica, ma rivela probabilmente un limite strutturale dei processi di popolamento di Wikidata.

```python
# filtro per periodo da analizzare
df_focus = df_cronologia[df_cronologia['data_pubblicazione'] >= 2000]
```

```python
# totale delle pubblicazioni in un anno
totale = df_focus['data_pubblicazione'].value_counts()
```

```python
# totale articoli che hanno DOI
con_doi = df_focus[df_focus['doi_disponibili'].notna()]['data_pubblicazione'].value_counts()
```

```python
#totale articoli che hanno argomenti
con_argomenti = df_focus[df_focus['argomenti'].notna()] ['data_pubblicazione'].value_counts()
```

```python
# raggruppamento per grafico
df_grafico = pd.DataFrame({
    'Totale Articoli': totale,
    'Articoli con DOI': con_doi,
    'Articoli con argomenti': con_argomenti
}).sort_index()
```

```python
ax = df_grafico.plot.line(
    marker=".",
    grid=True,
    color=["grey", "red", "blue"],
    figsize=(12, 6),
    title="Convergenza Qualitativa (2000-2025)",
    xlabel="Anno",
    ylabel="Numero di articoli")
```

Il grafico rivela chiaramente che la crescita del volume di pubblicazioni non è stata accompagnata da una crescita proporzionale della qualità dei metadati. Nonostante l'esplosione di record tra il 2010 e il 2020, la distanza tra la linea grigia e le linee blu e rosse indica che stiamo accumulando articoli censiti ma privi di **DOI** e **argomenti** che li renderebbero realmente utili per la ricerca automatizzata.

Anche se si potrebbe pensare il contrario, il divario non si riduce nei periodi più recenti. Ci aspetteremmo che le pubblicazioni nate nell'era digitale (post-2015) siano complete al 100%, invece il grafico mostra che la percentuale di **DOI** e **argomenti** mancanti rimane costante. Questo suggerisce che il caricamento su Wikidata non venga effettuato con il rigore necessario, ma in modo approssimativo.
